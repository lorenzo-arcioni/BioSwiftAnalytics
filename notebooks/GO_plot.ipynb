# Go2Nt Plotter

Get go mapper file

```python
from goatools.base import get_godag
godag = get_godag("go.obo")
```

## Creation of the dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

%matplotlib inline
```

```python
df = pd.read_csv("GO.emapper.annotations.tsv", sep='\t')

GOs = df.GOs

GOs.dropna(inplace=True)

GOs_dict = {}

for go in GOs:
    for x in go.split(","):
        if x in GOs_dict:
            GOs_dict[x] += 1
        else:
            GOs_dict[x] = 1
```

```python
print(GOs_dict["GO:0061306"])
```

```python
# Get GO terms selection
go_ids_selected = set(GOs_dict.keys())

from goatools.gosubdag.gosubdag import GoSubDag
gosubdag = GoSubDag(go_ids_selected, godag)

nts=[]

for go in go_ids_selected:
    try:
        nts.append(gosubdag.go2nt[go])
    except: 
        pass
```

```python
fmt_str = '{I:2}) {NS} {GO:10} {dcnt:11}        D{depth:02}  {GO_name}'

GOs_desc_dict = {}

# Print selected GO information
print('IDX NS GO ID      Descendants Count Depth Name')
print('--- -- ---------- ----------------- ----- --------------------')
for idx, nt_go in enumerate(sorted(nts, key=lambda nt: nt.depth), 1):
    if nt_go.GO_name in GOs_desc_dict:
        GOs_desc_dict[nt_go.GO_name] = [nt_go.NS, GOs_desc_dict[nt_go.GO_name][1] + GOs_dict[nt_go.GO]]
    else:
        GOs_desc_dict[nt_go.GO_name] = [nt_go.NS, GOs_dict[nt_go.GO]]
    #print(fmt_str.format(I=idx, **nt_go._asdict()))
```

```python
#Get GO info of labels that have more than 2 hits
GOs_info = [[x, GOs_desc_dict[x][0], GOs_desc_dict[x][1]] for x in GOs_desc_dict if GOs_desc_dict[x][1] > 1]

#Sorting and creating a list of three elements types (BP, CC, MF)
GOs_info.sort(key = lambda x: (x[1], x[2]))
GOs_info = [[x for x in GOs_info if x[1] == 'MF'], [x for x in GOs_info if x[1] == 'CC'], [x for x in GOs_info if x[1] == 'BP']]
#Create the barh plot
fig, ax = plt.subplots()

fig.set_size_inches(18.5, 10.5)

numbers_of_bars = min(len(min(GOs_info, key=lambda x: len(x))), 7)

X_axis = np.arange(numbers_of_bars)

labels = [k[0] for x in GOs_info for k in x[-numbers_of_bars:]]
items  = [k[2] for x in GOs_info for k in x[-numbers_of_bars:]]
colors = [k[1] for x in GOs_info for k in x[-numbers_of_bars:]]

for i in range(len(colors)):
    if colors[i] == 'BP':
        colors[i] = 'green'
    elif colors[i] == 'CC':
        colors[i] = 'red'
    else:
        colors[i] = 'blue'
        

y_pos = np.arange(len(items))

hbars = ax.barh(y_pos, items, align='center', color=colors)
ax.set_yticks(y_pos, labels=labels)

for t in range(len(ax.yaxis.get_ticklabels())):
    ax.yaxis.get_ticklabels()[t].set_color(colors[t])


ax.set_ylabel('Function name')
ax.set_xlabel('Number of genes')
ax.set_title('The most enriched GO terms')
red_patch = mpatches.Patch(color='red', label='Cellular component')
blue_patch = mpatches.Patch(color='blue', label='Molecular function')
green_patch = mpatches.Patch(color='green', label='Biological process')
ax.legend(handles=[blue_patch ,red_patch, green_patch])


for index, value in enumerate(items):
    plt.text(value, index, " " + str(value), va='center')

plt.savefig('GO_count.png', bbox_inches = 'tight', dpi=350)
plt.show()
```
